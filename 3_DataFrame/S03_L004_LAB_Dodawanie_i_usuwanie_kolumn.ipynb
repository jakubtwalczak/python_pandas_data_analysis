```python
import pandas as pd

fuel = pd.read_csv('fuel.csv', low_memory=False, usecols=['Vehicle ID', 'Year', 'Make', 'Model', 'Class', 'Fuel Type', 'Combined MPG (FT1)'])
fuel.head()
```

```python
fuel['Combined KPL'] = fuel['Combined MPG (FT1)'] * 1.609 / 3.78
fuel.head()
```

```python
fuel.insert(loc=5, column='liters per km', value=100 / fuel['Combined KPL'])
fuel.head()
```

```python
fuel.drop(columns='Combined MPG (FT1)', inplace=True)
fuel.head()
```

```python
del fuel['Combined KPL']
fuel.head()
```
