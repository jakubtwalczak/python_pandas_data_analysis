```python
import pandas as pd
import numpy as np

fuel = pd.read_csv('fuel.csv', usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)','Start Stop Technology'],
                   low_memory=False)
fuel.head()
```

```python
listOfCars = ['Renault', 'Toyota', 'Ford']
isInList = fuel['Make'].isin(listOfCars)
isInList.head()
```

```python
fuel[isInList]
```

```python
isNoStartStopDefined = fuel['Start Stop Technology'].isnull()
isNoStartStopDefined.head()
```

```python
fuel[isNoStartStopDefined]
```

```python
isStartStopDefined = ~fuel['Start Stop Technology'].isnull()
isStartStopDefined.head()
```

```python
fuel[isStartStopDefined]
```

```python
mpg50_60 = fuel['Combined MPG (FT1)'].between(50,60)
mpg50_60.head()
```

```python
fuel[mpg50_60]
```

```python
print(f"There are {len(fuel[isInList])} cars within the list {listOfCars}.")
print(f"There are {len(fuel[isNoStartStopDefined])} cars where Start/Stop technology is not determined.")
print(f"There are {len(fuel[isStartStopDefined])} cars where Start/Stop technology is determined.")
print(f"There are {len(fuel[mpg50_60])} cars with MPG value between 50 and 60.")
```
