```python
import pandas as pd
import numpy as np

fuel = pd.read_csv('fuel.csv', usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'], low_memory=False)
fuel.head()
```

```python
fuel.info(memory_usage='deep')
```

```python
fuel['Year'] = fuel['Year'].astype('int16')
fuel.info(memory_usage='deep')
```

```python
fuel['Make'].value_counts()
```

```python
fuel['Make'] = fuel['Make'].astype('category')
fuel.info(memory_usage='deep')
```

```python
for col in ['Model','Class','Fuel Type']:
    fuel[col] = fuel[col].astype('category')
fuel.info(memory_usage='deep')
```

```python
fuel['Combined MPG (FT1)'] = fuel['Combined MPG (FT1)'].astype('float16')
fuel.info(memory_usage='deep')
```
