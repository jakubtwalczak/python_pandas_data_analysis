```python
import pandas as pd
import numpy as np

fortune500 = pd.read_csv("Fortune_500_2017.csv", usecols=['Rank','Title','Employees','Profits','Assets'], index_col='Rank')
fortune500.head()
```

```python
fortune500['Employees'].rank(ascending=False).head()
```

```python
fortune500['RankByEmployee'] = fortune500['Employees'].rank(ascending=False)
fortune500.head()
```

```python
fortune500['Profits'].rank(ascending=False).head()
```

```python
fortune500['RankByProfits'] = fortune500['Profits'].rank(ascending=False)
fortune500.head()
```
