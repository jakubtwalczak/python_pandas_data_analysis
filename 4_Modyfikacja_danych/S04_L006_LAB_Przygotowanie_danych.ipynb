```python
columns_to_load = ["status_message", "status_type", "link_name", "num_reactions", "num_shares", "num_likes"]
fb = pd.read_csv('mrbean_facebook_statuses_with_nulls.csv', usecols=columns_to_load)
fb.head()
```

```python
fb.info(memory_usage='deep')
```

```python
len(fb)
```

```python
fb['status_type'].nunique()
```

```python
fb['status_type'].value_counts()
```

```python
fb['status_type'] = fb['status_type'].astype('category')
fb.info(memory_usage='deep')
```

```python
fb['link_name'].nunique()
```

```python
fb['link_name'].value_counts()
```

```python
fb['link_name'] = fb['link_name'].astype('category')
fb.info(memory_usage='deep')
```

```python
fb[['num_reactions', 'num_shares']] = fb[['num_reactions', 'num_shares']].fillna(0)
fb[['num_reactions', 'num_shares', 'num_likes']] = fb[['num_reactions', 'num_shares', 'num_likes']].astype('int32')
fb.info(memory_usage='deep')
```
