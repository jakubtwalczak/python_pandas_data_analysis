```python
import pandas as pd
import numpy as np

professions = pd.read_csv('Prestige.csv')
professions.head()
```

```python
professions = pd.concat([professions, pd.DataFrame({
          'name': ['data scientist'],
          'education': [np.nan],
          'income': [np.nan],
          'women': [np.nan],
          'prestige': [np.nan],
          'census': [np.nan],
          'type': [np.nan]})])
professions.tail()
```

```python
taxi = professions.loc[professions['name'] == 'taxi.drivers']
taxi
```

```python
professions.drop(taxi.index, inplace=True)
professions.tail()
```

```python
professions = pd.concat([professions, taxi])
professions.tail()
```

```python
del professions['census']
professions.head()
```

```python
professions.drop(columns='type', inplace=True)
professions.head()
```
