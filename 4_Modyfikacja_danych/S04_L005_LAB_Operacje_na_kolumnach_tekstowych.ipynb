```python
import pandas as pd
import numpy as np

fortune = pd.read_csv('Fortune_500_2017.csv', usecols=['Rank','Title','Industry','Hqlocation','Sector'])
fortune.head()
```

```python
fortune['Sector'] = fortune['Sector'].str.upper()
fortune.head()
```

```python
fortune[fortune['Industry'].str.lower().str.contains('comp')]
```

```python
fortune[['City', 'State']] = fortune['Hqlocation'].str.split(',', expand=True)
fortune.head()
```

```python
def build_shortcut(row):
    industry = row['Industry']
    result = ''
    for i in industry.split(' '):
        result += i[0]
    return result
```

```python
build_shortcut({'Industry': 'Factory Under Newspaper'})
```

```python
fortune['IndustryShort'] = fortune.apply(build_shortcut, axis=1)
fortune.head()
```
